```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
from selenium.webdriver.support import expected_conditions as EC
import time

extracted_text = []
aux_pages = 0
links = []
```

```python
# Golpe da troca de cartão
while aux_pages <= 2:
    offset = aux_pages
    url = f'https://search.brave.com/news?q=Golpe%20da%20troca%20de%20cart%C3%A3o&offset={offset}&spellcheck=0'
    print(f"\n📄 Coletando página {aux_pages + 1} (offset={offset})")

    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    main_div = soup.select_one('body > div > div:nth-of-type(1) > div:nth-of-type(4) > div > main > div')

    if main_div:
        a_tags = main_div.find_all('a', href=True)
        for i, a in enumerate(a_tags, 1):
            print(f"[{i}] {a['href']}")
            links.append(a['href'])
    else:
        print("❌ Div principal não encontrada.")

    aux_pages += 1
```
